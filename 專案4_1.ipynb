資料集一 MDS

```python
import numpy as np
from sklearn.manifold import MDS
import matplotlib.pyplot as plt

# 假設距離矩陣（單位：公里）
distance_matrix = np.array([
    [0, 61.43, 131.35, 178.32, 294.46, 183.63, 267.31],
    [61.43, 0, 79.39, 128.90, 249.16, 162.56, 234.82],
    [131.35, 79.39, 0, 48.86, 170.78, 109.48, 163.39],
    [178.32, 128.90, 48.86, 0, 121.63, 91.9, 122.82],
    [294.46, 249.16, 170.78, 121.63, 0, 137.74, 71.76],
    [183.63, 162.56, 109.48, 91.9, 137.74, 0, 86.72],
    [267.31, 234.82, 163.39, 122.82, 71.76, 86.72, 0]
])

# 使用MDS進行降維
mds = MDS(n_components=2, dissimilarity="precomputed", random_state=42)
pos = mds.fit_transform(distance_matrix)

# 火車站名稱
stations = ['台北', '新竹', '台中', '斗六', '高雄', '花蓮玉里', '台東知本']

# 定義顏色列表
colors = ['red', 'blue', 'green', 'purple', 'orange', 'brown', 'pink']

# 繪製MDS結果
plt.figure(figsize=(10, 8))
for i, (x, y) in enumerate(pos):
    plt.scatter(x, y, color=colors[i], label=stations[i])
    plt.annotate(stations[i], (x, y))

plt.title('台灣火車站MDS圖')
plt.xlabel('維度1')
plt.ylabel('維度2')
plt.legend()
plt.grid()
plt.show()
```
